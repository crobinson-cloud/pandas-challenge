```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on="school_name")
school_data_complete.head()
```

## District Summary

```python
school_data_complete["school_name"].unique()
```

```python
# Calculate the total number of unique schools - use Len() on the list of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students - count of the number of student Ids
student_count = school_data_complete["Student ID"].count()
student_count
```

```python
# Look at the school data dataframe - take a sum of the budget column
school_data
```

```python
# Calculate the total budget 
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score - take the mean of the math_score column from data complete dataframe
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score - take the mean of the reading_score column from the data complete dataframe
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passMathStudents = school_data_complete[school_data_complete["math_score"]>= 70]
passMathStudents.head()
numPassMath = len(passMathStudents)
numPassMath
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
# (number of students who passed math / total number of students) * 100
percentagePassedMath =(numPassMath / student_count ) * 100
percentagePassedMath
```

```python
# use conditional to find all students who had a score of 70 or higher in reading
passReadingStudents = school_data_complete[school_data_complete["reading_score"]>= 70]
#passReadingStudents.head()
numPassReading = len(passReadingStudents)
numPassReading
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
# (number of students who passed reading / total number of students) * 100
percentagePassedReading = (numPassReading / student_count) * 100
percentagePassedReading 
```

```python
# use conditional to find all students who had a score of 70 or higher in math
passingStudents = school_data_complete[(school_data_complete["math_score"]>= 70) & (school_data_complete["reading_score"]>= 70)]
#passingStudents.head()
numPassingStudents = len(passingStudents)
numPassingStudents
```

```python
# Use the following to calculate the percentage of students that passed math and reading
# (number of students who passed math and reading / total number of students) * 100
percentagePassed = (numPassingStudents / student_count) * 100
percentagePassed
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame - make a dictionary of lists
districtSummary = pd.DataFrame(
    {
        "Total Schools": [school_count],
        "Total Students": [student_count],
        "Total Budget": [total_budget],
        "Average Math Score": [average_math_score],
        "Average Reading Score": [average_reading_score],
        "% Passing Math": [percentagePassedMath],
        "% Passing Reading": [percentagePassedReading],
        "% Overall Passing": [percentagePassed]
    }
)
# Formatting
districtSummary["Total Students"] = districtSummary["Total Students"].map("{:,}".format)
districtSummary["Total Budget"] = districtSummary["Total Budget"].map("{:,.2f}".format)
districtSummary["Average Math Score"] = districtSummary["Average Math Score"].map("{:.2f}".format)
districtSummary["Average Reading Score"] = districtSummary["Average Reading Score"].map("{:.2f}".format)
districtSummary[" % Passing Math"] = districtSummary["% Passing Math"].map("{:.2f}%".format)
districtSummary[" % Passing Reading"] = districtSummary["% Passing Reading"].map("{:.2f}%".format)
districtSummary[" % Overall Passing"] = districtSummary["% Overall Passing"].map("{:.2f}%".format)


# Display the DataFrame
districtSummary
```
